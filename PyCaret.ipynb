## Paramater Setting for Clustering Algorithms

```python
data=pd.read_csv('/content/sample_data/Iris.csv')
data.head()
```

```python
from pycaret.clustering import *

def modelfunction(clustering_algo='kmeans',transformation = False, normalize = False, pca = False,
      normalize_method = 'zscore',
      transformation_method = 'yeo-johnson',
      pca_method = 'linear', verbose=False):

  setup(data = data, transformation = transformation, normalize = normalize, pca = pca,
      normalize_method = 'zscore',
      transformation_method = 'yeo-johnson',
      pca_method = 'linear', verbose=False)

  print("For Cluster = 3")
  x = create_model(clustering_algo, num_clusters = 3)
  metrics1=pull()

  print("For Cluster = 4")
  x = create_model(clustering_algo, num_clusters = 4)
  metrics2=pull()

  print("For Cluster = 5")
  x = create_model(clustering_algo, num_clusters = 5)
  metrics3=pull()

  # header = pd.MultiIndex.from_product([['No Data PreProcessing'],
  #                                   ['Cluster1','Cluster2','Cluster3']],
  #                                     names=['loc','S']
  #                                   )

  newtable=pd.concat([metrics1,metrics2,metrics3],axis=0)
  newtable=newtable.T
  columnname=['Cluster1','Cluster2','Cluster3']
  newtable.columns=columnname



  print(newtable)
  # df = pd.DataFrame(newtable,
  #                 columns=header)
  # print(df)

  return newtable
```

```python
cluster_algo=['kmeans','meanshift','hclust']

list=[]

for i in range(len(cluster_algo)):

  newtable=pd.DataFrame()

  keys=['With No Data PreProcessing','Normalization','Transformation','PCA','Normalize+Tranforme','Normalize+Transform+PCA']


  a=modelfunction(clustering_algo=cluster_algo[i],normalize=True)
  # With Normalization
  b=modelfunction(clustering_algo=cluster_algo[i])
  # With Transformation
  c=modelfunction(clustering_algo=cluster_algo[i],transformation = True)
  # wITH pca
  d=modelfunction(clustering_algo=cluster_algo[i],pca = True)
  # With N+T
  e=modelfunction(clustering_algo=cluster_algo[i],normalize=True,transformation = True)
  # with T+N+PCA
  f=modelfunction(clustering_algo=cluster_algo[i],normalize=True,transformation = True)

  newtable=pd.concat([a,b,c,d,e,f],axis=1,keys=keys)
  list.insert(i,newtable)

```

```python
Cluster_names=['Kmeans Clustering','Kmeans Shift Clustering','Heirarchial Clustering']
i=0
# excel_file_path = f"data{i}.xlsx"

with pd.ExcelWriter('result.xlsx') as writer:
  for i, df in enumerate(list):
      print(f"Clustering Algo {i+1}:")
      print(Cluster_names[i])
      print()
      df.to_excel(writer, sheet_name=Cluster_names[i])

print()
print("All Model Result are successfully imported in Excel File with result of each Clustering Algo on Separate Sheet")
```
